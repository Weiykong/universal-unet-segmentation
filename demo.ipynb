```python
# 3. Create a Dummy Image for Testing
import numpy as np
import tifffile
import os

img = np.zeros((512, 512), dtype=np.float32)
y, x = np.ogrid[:512, :512]
center = (256, 256)
mask = ((y - center[0])**2 + (x - center[1])**2) < 50**2
img[mask] = 1.0
img += np.random.normal(0, 0.1, img.shape)

os.makedirs('data/inference_input', exist_ok=True)
tifffile.imwrite('data/inference_input/test_image.tif', img.astype(np.float32))
print("Created synthetic test image.")
```

```python
# 4. Run Inference
!python run.py --mode inference
```

```python
# 5. Visualize Result
import matplotlib.pyplot as plt
import tifffile

raw = tifffile.imread('data/inference_input/test_image.tif')
prob = tifffile.imread('output/prob_test_image.tif')

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(raw, cmap='gray')
ax[0].set_title('Input Raw')
ax[1].imshow(prob, cmap='magma')
ax[1].set_title('Probability Map Output')
plt.show()
```
